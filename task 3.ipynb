```python
import pandas as pd

# Step 1: Create sample stock data
data = {
    'Date': pd.date_range(start='2025-11-11', periods=15, freq='D'),
    'Close': [100, 102, 101, 105, 107, 110, 108, 111, 115, 117, 120, 118, 121, 123, 125]
}

df = pd.DataFrame(data)

# Step 2: Calculate 5-day Simple Moving Average
df['SMA_5'] = df['Close'].rolling(window=5).mean()

# Step 3: Display DataFrame
print(df)
```

```python
#create simple moving average in dataframe and do not use predefined functionality use your own logic.
import pandas as pd

# Step 1: Create sample stock data
data = {
    'Date': pd.date_range(start='2025-11-01', periods=10, freq='D'),
    'Close': [100, 102, 101, 105, 107, 110, 108, 111, 115, 117]
}

df = pd.DataFrame(data)

# Step 2: Define window size for SMA
window = 5
sma_values = []

# Step 3: Calculate SMA manually using loops
for i in range(len(df)):
    if i + 1 < window:
        # Not enough data points for SMA
        sma_values.append(None)
    else:
        # Take the last 'window' closing prices
        total = 0
        for j in range(i - window + 1, i + 1):
            total += df.loc[j, 'Close']
        sma = total / window
        sma_values.append(sma)

# Step 4: Add new column to DataFrame
df['SMA_5'] = sma_values

# Step 5: Display the result
print(df)
```
